#### 1. Using a line plot visualize the Total profit data provided for each month.

**Note:**It must be implemented in three methods
- Generated line plot must include the following properties:

X axis name = Month Number <br>
Y axis name = Total profit <br>
The line plot graph should look like this.


```python
####################################### Line Plot using Matplotlib visualization #########################################
```

```python
# Create a line plot with formatting options
plt.plot(df['month_number'], df['total_profit'], color='blue')

# set title
plt.title('Company profit per month')

# Set x and y axis labels
plt.xlabel('Month Number')
plt.ylabel('Total Profit')

# Set x ticks
plt.xticks(df['month_number'])
plt.xticks(range(1, 13))
plt.yticks(range(100000, 600000,100000))

# Show the plot
plt.show()
```

#### 3. Read face cream and facewash product sales data and show it using the bar chart.


- The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.  
- **Note:**It must be implemented in three methods  

The bar chart should look like this.


```python
#################################### bar Plot using Matplotlib visualization #########################################
```

```python
# create a figure and axis
fig, ax = plt.subplots(figsize=(4,3))

# specify the width of each bar
bar_width = 0.35

# get x data for each bar
points1 = df['month_number']
points2 = df['month_number'] + bar_width

# get y data for each bar
frequency1 = df['facecream']
frequency2 = df['facewash']

# plot the bars side by side
ax.bar(points1, frequency1, width=bar_width, label='Face Cream')
ax.bar(points2, frequency2, width=bar_width, label='Face Wash')

# set x and y labels
ax.set_xlabel('Month Number')
ax.set_ylabel('Sales units in number')

# set the plot title
ax.set_title('FaceWash and FaceCream Sales data')

# add a legend
ax.legend()

# draw vertical lines
for i in range(1, 13):
    plt.axvline(i, color='grey', linestyle='--')

# draw horizontal lines
for i in range(0, 4000, 500):
    plt.axhline(i, color='grey', linestyle='--')


# Set x ticks
plt.xticks(range(1, 13))

# show the plot
plt.show()
```

#### 5. Read the total profit of each month and show it using the histogram to see the most common profit ranges.  
**Note:**It must be implemented in three methods.  
The histogram should look like this.


```python
####################################### histogram Plot using Matplotlib visualization #########################################
```

```python
# create figure and axis
fig, ax = plt.subplots(figsize=(4,3))

# plot histogram without border
ax.hist(df['total_profit'], edgecolor="none", color="b", orientation='vertical', linewidth=0)

# set x and y labels
ax.set_xlabel('Profit range in dollar')
ax.set_ylabel('Actual profit in dollar')

# set the plot title
ax.set_title('Profit Data')

# set x and y limits and ticks
ax.set_xlim(150000, 350000)
ax.set_xticks(range(150000, 350001, 25000))
ax.set_ylim(0, 6)
ax.set_yticks(range(0, 6, 1))

# show the plot
plt.show()
```

```python
####################################### histogram  Plot using Seaborn visualization #########################################
```

```python
#create figure and axis
fig, ax = plt.subplots(figsize=(5,3))

# plot histogram without border
sns.histplot(data=df, x='total_profit', ax=ax, color="b", edgecolor="none")

# set x and y labels
ax.set_xlabel('Profit range in dollar')
ax.set_ylabel('Actual profit in dollar')

# set the plot title
ax.set_title('Profit Data')

# set x and y limits and ticks
ax.set_xlim(150000, 350000)
ax.set_xticks(range(150000, 350001, 25000))
ax.set_ylim(0, 6)
ax.set_yticks(range(0, 6, 1))

# show the plot
plt.show()
```

```python
#################################### histogram Plot using Pandas visualization #####################################
```

#### 6. Calculate total sale data for last year for each product and show it using a Pie chart.
- **Note:**
    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python
#################################### pie Plot using Matplotlib visualization #####################################
```

```python
# calculate total sales for each product
total_sales = df.iloc[:, 1:-2].sum()

# create a figure and axis
fig, ax = plt.subplots(figsize=(3,3))

# create a pie chart
ax.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%')

# set the chart title
ax.set_title('Sale Data')

# show the chart
plt.show()
```
